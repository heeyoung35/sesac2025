### TTS(Text-to-Speech)
텍스트를 자연스러운 음성으로 변환하는 AI 모델이다.

** 음성 선택지 **
- ash
- ballad
- sage
- marin
- cedar
- fable
- coral

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from openai import OpenAI

client = OpenAI()

input = "내년 서울 아파트 입주 물량이 올해의 절반 수준으로 떨어질 전망이다. 전세 매물 감소와 전세의 월세화 등으로 임차인의 주거 비용이 올라가는 상황에서 내년 임차인의 주거 부담은 더욱 커질 것으로 보인다."

# 실시간 스트리밍을 사용해야할 때 with_streaming_response을 사용해야 한다. 
with client.audio.speech.with_streaming_response.create(
    model="tts-1",
    voice="nova",
    input=input[:200]
) as response:
    response.stream_to_file("output.mp3")
    print("파일 생성이 완료되었습니다.")

```
