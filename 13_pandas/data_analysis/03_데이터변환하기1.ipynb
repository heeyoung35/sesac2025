### 데이터 변환하기1

```python
import pandas as pd

# 파일 불러오기
df=pd.read_csv('./data/seoul_park.csv')
df.head()
```

```python
df
```

```python
df.info()   #데이터타입에 object는 텍스트이다. 텍스트로는 통계값을 구할 수 없다. 데이터를 변환해서 정수형으로 변환한다.  숫자만 남기고 모두 제거한다.
```

```python
# 텍스트 데이터를 정수형으로 바꾸기 전에 텍스트에서 숫자만 남기고 제거한다.
# 이를 위해 숫자로 바꾸어야 하는 컬럼들에서 replace()를 활용하여 -를 0으로 바꾸고 콤마를 제거한다. 
# 데이터 타입 변경전 전처리 과정

# str.replace() 반복문을 통해 접근해서 불필요한 것을 제거한다.  
columns=['유료합계', '어른', '청소년', '어린이', '외국인', '단체', '무료합계', '총계']
for i in columns:
    df[i]=df[i].str.replace('-', '0')  # -을 replace로 0으로 교제하겠다. 원본이 아니라 새로운 데이터를 만들어서 원본에 반영하고 싶다면 =를 통해서 덮어 씌우는 것이다. 
    df[i]=df[i].str.replace(',','')
```

```python
df.info()   # 여전히 데이터 타입은 object이다. 
```

```python
df.info()
```

```python
df["어른"]=df["어른"].astype(int)
# 데이터프레임에 변형을 가하는 것은 원본데이터를 바로 바꾸지 않기 때문에 원본에 반영하겠다고 해야 원본데이터에 반영이 된다. 

df.info()
```

### 데이터 타입 변환: to_numeric()

```python
# 숫자로만 변환 가능
df["유료합계"]=pd.to_numeric(df["유료합계"])

df.info()
```

```python
## df의 "외국인" 컬럼의 자료형을 숫자형으로 변환해 변수 foreigner애 저장하세요
foreigner=df["외국인"].astype(int)

foreigner.info()
```

```python
columns=['유료합계', '어른', '청소년', '어린이', '외국인', '단체', '무료합계', '총계']
for i in columns:
    df[i]=pd.to_numeric(df[i])

    df.info()
```

### 데이터 타입 변화: to_datetime()

```python
df["날씨"]=pd.to_datetime(df["날짜"])
df.info()
```

```python
# datetiome 형태의 데이터에는 연, 월, 일 정보가 담겨있어 이를 활용하여 새로운 정보를 생성할 수 있다.
df['연']=df['날짜'].dt.year
df['월']=df['날짜'].dt.month
df['일']=df['날짜'].dt.day
df['요일']=df['날짜'].dt.dayofweek  # 요일은 숫자로 저장(월~일:0~6)
```
